```python
# ============================
# Credit Card Users Churn Prediction + SHAP & LIME
# ============================

###############################################################################
# 1. Install & Import Libraries
###############################################################################
# Installing required libraries (specified versions)
!pip install pandas==2.2.2 numpy==1.26.4 scikit-learn==1.3.1 matplotlib==3.8.0 seaborn==0.13.1 \
   imbalanced-learn==0.11.0 xgboost==2.0.3 shap==0.43.0 lime==0.2.0.1 scikit-optimize -q

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings("ignore")

# Data handling & utilities
import pandas as pd
import numpy as np
import joblib

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Modeling
import sklearn
from sklearn.model_selection import (
    train_test_split, RandomizedSearchCV, StratifiedKFold
)
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, classification_report, roc_auc_score,
    confusion_matrix, roc_curve, recall_score, precision_score, f1_score
)

# Imbalanced handling
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Advanced optimization
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

# Neural networks (Keras/TensorFlow)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Interpretability
import shap
import lime
import lime.lime_tabular

# Print library versions for reference
print("Package Versions:")
print(f"Pandas: {pd.__version__}")
print(f"Numpy: {np.__version__}")
print(f"Scikit-learn: {sklearn.__version__}")
print(f"SHAP: {shap.__version__}")
print()


###############################################################################
# 2. Load & Explore Data (EDA)
###############################################################################

# 2.1 Data Preprocessing
from google.colab import drive
drive.mount('/content/drive')  # Mount Google Drive (Colab environment)

# Read CSV data
df = pd.read_csv('/content/drive/MyDrive/BankChurners.csv')
print("\nInitial Data Overview:")
print(df.info())
print(df.head())

# Check for duplicates
print("\nNumber of Duplicate Rows:", df.duplicated().sum())

# Drop CLIENTNUM if present (just an identifier)
if 'CLIENTNUM' in df.columns:
    print("Unique CLIENTNUM records:", df['CLIENTNUM'].nunique())
    df.drop(columns=['CLIENTNUM'], inplace=True)

# Convert 'Attrition_Flag' -> numeric (0 = Existing, 1 = Attrited)
df['Attrition_Flag'] = df['Attrition_Flag'].apply(lambda x: 1 if x == "Attrited Customer" else 0)

# Check missing values
print("\nMissing Value Counts:")
print(df.isnull().sum())

# Quick distribution checks of relevant categorical columns
if 'Education_Level' in df.columns:
    print("\nEducation_Level distribution (including 'Unknown'):")
    print(df['Education_Level'].value_counts(dropna=False))

if 'Income_Category' in df.columns:
    print("\nIncome_Category distribution (including 'Unknown'):")
    print(df['Income_Category'].value_counts(dropna=False))

# ----------------------------------------------------------------------------
# Helper Functions for EDA
# ----------------------------------------------------------------------------
def histogram_boxplot(data, feature, figsize=(12, 7), kde=False, bins=None):
    """
    Creates a combined boxplot & histogram for visualizing the distribution
    and outliers of a numeric feature.
    """
    fig, (ax_box, ax_hist) = plt.subplots(
        nrows=2, sharex=True,
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize
    )
    # Boxplot
    sns.boxplot(data=data, x=feature, ax=ax_box, showmeans=True, color="violet")
    # Histogram
    if bins:
        sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist, bins=bins, color="steelblue")
    else:
        sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist, color="steelblue")
    # Reference lines
    ax_hist.axvline(data[feature].mean(), color="green", linestyle="--", label="Mean")
    ax_hist.axvline(data[feature].median(), color="black", linestyle="-", label="Median")
    ax_hist.legend()
    plt.show()

def labeled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with annotated counts or percentages on each bar.
    """
    total = len(data[feature])
    count = data[feature].nunique()
    plt.figure(figsize=(n + 1, 5) if n else (count + 1, 5))
    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(data=data, x=feature, palette="Paired",
                       order=data[feature].value_counts().index[:n].sort_values())
    for p in ax.patches:
        label = "{:.1f}%".format(100 * p.get_height() / total) if perc else p.get_height()
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(label, (x, y), ha="center", va="center", size=12,
                    xytext=(0, 5), textcoords="offset points")
    plt.show()

def stacked_barplot(data, predictor, target):
    """
    Shows category counts & stacked bar chart of predictor vs. target distribution.
    """
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(by=sorter, ascending=False)
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(by=sorter, ascending=False)
    tab.plot(kind="bar", stacked=True, figsize=(count + 1, 5))
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()

def distribution_plot_wrt_target(data, predictor, target):
    """
    Plots histograms & boxplots of a numeric predictor grouped by the target label.
    """
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    target_vals = data[target].unique()

    # Histograms for each target category
    sns.histplot(data=data[data[target] == target_vals[0]], x=predictor, kde=True, ax=axs[0, 0], color="teal")
    axs[0, 0].set_title(f"Distribution (target={target_vals[0]})")

    sns.histplot(data=data[data[target] == target_vals[1]], x=predictor, kde=True, ax=axs[0, 1], color="orange")
    axs[0, 1].set_title(f"Distribution (target={target_vals[1]})")

    # Boxplots with and without outliers
    sns.boxplot(data=data, x=target, y=predictor, ax=axs[1, 0], palette="gist_rainbow")
    axs[1, 0].set_title("Boxplot w.r.t target")

    sns.boxplot(data=data, x=target, y=predictor, showfliers=False, ax=axs[1, 1], palette="gist_rainbow")
    axs[1, 1].set_title("Boxplot (no outliers) w.r.t target")

    plt.tight_layout()
    plt.show()

# ----------------------------------------------------------------------------
# 2.2 Feature Distribution Analysis
# ----------------------------------------------------------------------------
print("\nAttrition Flag Distribution:")
print(df['Attrition_Flag'].value_counts(normalize=True)*100)

num_features = ['Credit_Limit','Avg_Utilization_Ratio','Total_Trans_Amt','Total_Trans_Ct']
print("\nNumeric Feature Descriptive Stats:")
print(df[num_features].describe())

cat_features = ['Income_Category','Education_Level','Marital_Status']
for c in cat_features:
    if c in df.columns:
        print(f"\nCategorical Distribution: {c}")
        print(df[c].value_counts())

# ----------------------------------------------------------------------------
# 2.3 Correlation Analysis
# ----------------------------------------------------------------------------
df_numeric = df.select_dtypes(include=[np.number])
corr = df_numeric.corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric-Only Columns)")
plt.show()

# Convert Gender to numeric
gender_map = {'M': 1, 'F': 0}
if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].map(gender_map)

###############################################################################
# 3. Feature Engineering
###############################################################################

# 3.1 Handling Categorical
if 'Card_Category' not in df.columns:
    df['Card_Category'] = 'Unknown'
df = pd.get_dummies(df, columns=['Card_Category'], drop_first=False)

# Map known categories in Education / Income
education_map = {
    'Uneducated': 0,'High School':1,'College':2,'Graduate':3,'Post-Graduate':4,'Doctorate':5,'Unknown':-1
}
if 'Education_Level' in df.columns:
    df['Education_Level'] = df['Education_Level'].map(education_map)

income_map = {
    'Less than $40K': 0,'$40K - $60K': 1,'$60K - $80K': 2,
    '$80K - $120K': 3,'$120K +': 4,'Unknown': -1
}
if 'Income_Category' in df.columns:
    df['Income_Category'] = df['Income_Category'].map(income_map)

# Encode remaining object columns
obj_cols = df.select_dtypes(include=['object']).columns
for col in obj_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# 3.2 Creating New Features
df['Credit_Usage_Ratio'] = df['Total_Revolving_Bal'] / df['Credit_Limit']
df['Activity_Ratio'] = df['Contacts_Count_12_mon'] / df['Months_on_book']
df['Transaction_Change_Rate'] = df['Total_Ct_Chng_Q4_Q1'] / df['Months_on_book']
df['Loyalty_Metric'] = df['Total_Relationship_Count'] / df['Months_on_book']
df['Avg_Transactions_Active_Month'] = df['Total_Trans_Ct'] / (df['Months_on_book'] - df['Months_Inactive_12_mon'])
df['Avg_Contacts_Active_Month'] = df['Contacts_Count_12_mon'] / (df['Months_on_book'] - df['Months_Inactive_12_mon'])

# Re-check correlation with the newly engineered features
plt.figure(figsize=(14, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix (Post-Engineering)")
plt.show()

###############################################################################
# Sample Visualization: Average Open-to-Buy by Income Category (Example Only)
###############################################################################
sample_data = {
    'IncomeCategory': ['Low', 'Low', 'Medium', 'Medium', 'High', 'High', 'High'],
    'OpenToBuy': [1000, 1100, 2500, 2600, 4000, 4200, 4100]
}
sample_df = pd.DataFrame(sample_data)

grouped = sample_df.groupby('IncomeCategory')['OpenToBuy'].agg(['mean', 'std']).reset_index()
plt.figure(figsize=(10, 6))
plt.bar(grouped['IncomeCategory'], grouped['mean'], yerr=grouped['std'],
        capsize=5, color='skyblue', edgecolor='black')
plt.xlabel('Income Category')
plt.ylabel('Average Open to Buy (Mean)')
plt.title('Average Open to Buy vs. Income Category (Sample Data)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

sns.barplot(x='IncomeCategory', y='OpenToBuy', data=sample_df, ci='sd',
            palette='pastel', edgecolor='black')
plt.xlabel('Income Category')
plt.ylabel('Average Open to Buy (Mean)')
plt.title('Average Open to Buy vs. Income Category w/ Error Bars')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Further correlation checks
df_numeric = df.select_dtypes(include=[np.number]).copy()
corr_matrix = df_numeric.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features (Final)")
plt.show()

# Flatten correlation matrix & show high correlations
corr_pairs = (
    corr_matrix.unstack()
    .reset_index()
    .rename(columns={'level_0': 'Feature1', 'level_1': 'Feature2', 0: 'Correlation'})
)
corr_pairs = corr_pairs[corr_pairs['Feature1'] < corr_pairs['Feature2']]
threshold = 0.70
high_corr = corr_pairs[corr_pairs['Correlation'].abs() > threshold]
print("Highly Correlated Feature Pairs (|corr| > 0.70):")
print(high_corr.sort_values(by='Correlation', key=lambda x: x.abs(), ascending=False))

###############################################################################
# Some Additional EDA / Visualization
###############################################################################
churn_by_income = df.groupby('Income_Category')['Attrition_Flag'].mean().reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='Income_Category', y='Attrition_Flag', data=churn_by_income, palette='viridis')
plt.title("Churn Rate by Income Category")
plt.xlabel("Income Category")
plt.ylabel("Churn Rate")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()

###############################################################################
# 4. Model Selection & Training
###############################################################################

# 4.1 Split Data (e.g., 70% train + val, 30% test)
X = df.drop(columns=['Attrition_Flag'])
y = df['Attrition_Flag']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# Further split train into train/val
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, stratify=y_train, random_state=42
)
print("\nData Splits:")
print("Train:", X_train.shape,
      "Val:", X_val.shape,
      "Test:", X_test.shape)

# 4.2 Impute & Handle Imbalance (SMOTE)
imputer = SimpleImputer(strategy='mean')
X_train_imputed = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_val_imputed   = pd.DataFrame(imputer.transform(X_val),     columns=X_val.columns,   index=X_val.index)
X_test_imputed  = pd.DataFrame(imputer.transform(X_test),    columns=X_test.columns,  index=X_test.index)

sm = SMOTE(random_state=42)
X_train_smote, y_train_smote = sm.fit_resample(X_train_imputed, y_train)
print("\nClass Distribution After SMOTE:")
print(y_train_smote.value_counts())

###############################################################################
# Logistic Regression (Baseline)
###############################################################################
log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train_smote, y_train_smote)

# Validation performance
y_pred_val_log = log_reg.predict(X_val_imputed)
val_acc_log = accuracy_score(y_val, y_pred_val_log)
val_auc_log = roc_auc_score(y_val, y_pred_val_log)
print(f"\nLogistic Regression Validation Accuracy: {val_acc_log:.3f}, AUC: {val_auc_log:.3f}")

# Test performance
y_pred_test_log = log_reg.predict(X_test_imputed)
test_acc_log = accuracy_score(y_test, y_pred_test_log)
test_auc_log = roc_auc_score(y_test, log_reg.predict_proba(X_test_imputed)[:, 1])
print(f"Logistic Regression Test Accuracy: {test_acc_log:.3f}, AUC: {test_auc_log:.3f}")
print("\nClassification Report (Logistic Regression):")
print(classification_report(y_test, y_pred_test_log))

###############################################################################
# 4.3 Random Forest (Initial)
###############################################################################
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_smote, y_train_smote)

# Validation performance
val_pred = rf.predict(X_val_imputed)
val_acc = accuracy_score(y_val, val_pred)
val_auc = roc_auc_score(y_val, val_pred)
print(f"\nRandom Forest Validation Accuracy: {val_acc:.3f}, AUC: {val_auc:.3f}")

# Test performance
test_pred = rf.predict(X_test_imputed)
test_acc = accuracy_score(y_test, test_pred)
test_auc = roc_auc_score(y_test, test_pred)
print(f"Random Forest Test Accuracy: {test_acc:.3f}, AUC: {test_auc:.3f}")

# Example: Plot random forest feature importance
# (Assume best_model is assigned to rf or a refined model)
# If you have a 'best_model' from tuning:
#importances = best_model.feature_importances_
#importances = rf.feature_importances_  # if we're just using rf
try:
    importances = rf.feature_importances_
    feature_names = X_train_imputed.columns
    indices = np.argsort(importances)[::-1]

    plt.figure(figsize=(10, 6))
    sns.barplot(x=importances[indices], y=feature_names[indices], palette='viridis')
    plt.title("Feature Importance (Random Forest)")
    plt.xlabel("Importance Score")
    plt.ylabel("Feature")
    plt.show()
except:
    pass  # If we didn't define 'best_model' yet

###############################################################################
# 4.3 Neural Network (Keras)
###############################################################################
def create_model(input_dim):
    """
    Simple feedforward neural network for binary classification.
    """
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(16, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(1, activation='sigmoid')  # single-unit output for churn (0/1)
    ])
    model.compile(
        loss='binary_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

# Convert to NumPy arrays
X_train_np = X_train_imputed.values
X_val_np   = X_val_imputed.values
y_train_np = y_train.values
y_val_np   = y_val.values

nn_model = create_model(input_dim=X_train_np.shape[1])
history = nn_model.fit(
    X_train_np, y_train_np,
    validation_data=(X_val_np, y_val_np),
    epochs=20,
    batch_size=64,
    verbose=1
)

# Evaluate on test set
X_test_np = X_test_imputed.values
y_test_np = y_test.values

y_pred_proba = nn_model.predict(X_test_np)
y_pred = (y_pred_proba > 0.5).astype(int).ravel()

acc = accuracy_score(y_test_np, y_pred)
auc = roc_auc_score(y_test_np, y_pred_proba)
print(f"\nNeural Network Test Accuracy: {acc:.3f}")
print(f"Neural Network Test ROC-AUC: {auc:.3f}")
print("Classification Report:\n", classification_report(y_test_np, y_pred))

###############################################################################
# 4.4 Hyperparameter Tuning (Random Search & Bayesian Optimization)
###############################################################################

# RandomizedSearch
param_grid = {
    "n_estimators": [50, 100, 150],
    "max_features": ['sqrt', 'log2'],
    "min_samples_leaf": [1, 2, 3]
}
tuned_rf = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=5, cv=3, scoring='roc_auc',
    random_state=42, n_jobs=-1, verbose=2
)
tuned_rf.fit(X_train_smote, y_train_smote)
best_model = tuned_rf.best_estimator_
print("\nBest Hyperparameters (RandomSearch):", tuned_rf.best_params_)
print("Best CV Score:", tuned_rf.best_score_)

# Bayesian Optimization
search_space = {
    'n_estimators': Integer(50, 300),
    'max_features': Categorical(['sqrt','log2',None]),
    'min_samples_leaf': Integer(1, 10),
    'max_depth': Integer(2, 20)
}
bayes_cv = BayesSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    search_spaces=search_space,
    n_iter=20,
    cv=StratifiedKFold(n_splits=3),
    scoring='roc_auc',
    n_jobs=-1,
    random_state=42,
    verbose=2
)
bayes_cv.fit(X_train_smote, y_train_smote)
print("Best params (Bayesian):", bayes_cv.best_params_)
print("Best ROC-AUC score:", bayes_cv.best_score_)

best_bayes_model = bayes_cv.best_estimator_

y_pred_test = best_bayes_model.predict(X_test_imputed)
test_acc = accuracy_score(y_test, y_pred_test)
test_auc = roc_auc_score(y_test, best_bayes_model.predict_proba(X_test_imputed)[:,1])
print(f"\nBayesian-Optimized RF -> Test Accuracy: {test_acc:.3f}, AUC: {test_auc:.3f}")

###############################################################################
# 5. Model Evaluation (Test Set) + SHAP & LIME
###############################################################################

# Evaluate final chosen model on test data
final_pred = best_model.predict(X_test_imputed)
final_acc = accuracy_score(y_test, final_pred)
final_auc = roc_auc_score(y_test, final_pred)
print(f"\nFinal Model on Test Set -> Accuracy: {final_acc:.3f}, AUC: {final_auc:.3f}")
print("\nClassification Report:")
print(classification_report(y_test, final_pred))

# Confusion Matrix
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, final_pred), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix (Test Data)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# (A) SHAP Analysis
explainer = shap.TreeExplainer(best_model)
X_test_shap = X_test_imputed[X_train_smote.columns]  # ensure matching columns

shap_values = explainer.shap_values(X_test_shap)
shap.summary_plot(shap_values, X_test_shap, plot_type="bar")  # summary bar plot

# If desired, a dependence plot on "Credit_Limit"
if 'Credit_Limit' in X_test_shap.columns:
    shap.dependence_plot("Credit_Limit", shap_values[1], X_test_shap)

# (B) LIME Analysis
feature_names = X_test_imputed.columns.tolist()
class_names = ["Non-Churn (0)", "Churn (1)"]

explainer_lime = lime.lime_tabular.LimeTabularExplainer(
    training_data=X_train_smote.values,
    feature_names=feature_names,
    class_names=class_names,
    verbose=True,
    mode='classification'
)

def predict_proba_fn(data_array):
    return best_model.predict_proba(data_array)

test_instance_idx = 0
instance = X_test_imputed.iloc[test_instance_idx].values
lime_exp = explainer_lime.explain_instance(
    data_row=instance,
    predict_fn=predict_proba_fn,
    num_features=5
)
lime_exp.show_in_notebook(show_table=True)
print("\nLIME Explanation as (Feature, Weight):", lime_exp.as_list())

###############################################################################
# Deployment & Business Strategy
###############################################################################

# 1. Save the final trained model
joblib.dump(best_model, "churn_rf_model.pkl")

# 2. Example conditions to offer retention strategies (pseudo-code)
churn_score = 0.75  # Hypothetical predicted churn probability
Income_Category = 1
Avg_Utilization_Ratio = 0.6
Months_Inactive_12_mon = 4

if (churn_score > 0.7) and (Income_Category <= 1):
    # Provide fee waiver or partial discount
    print("Offering fee waiver or partial discount.")

if (churn_score > 0.7) and (Avg_Utilization_Ratio > 0.5):
    # Offer credit limit increase
    print("Offering credit limit increase.")

if (churn_score > 0.7) and (Months_Inactive_12_mon >= 3):
    # Offer loyalty rewards or points
    print("Offering loyalty rewards or points.")


###############################################################################
# Project Insights / Conclusion
###############################################################################
print("""



1. Exploratory Data Analysis (EDA)

When I first loaded the Thera Bank churn dataset, I noticed it had 10,127 records and 21 columns. I used commands like df.info() and df.head() to check data types and see sample rows. I was happy to find a broad range of features, covering demographics (like Customer_Age and Gender), card usage metrics (like Total_Trans_Amt and Avg_Utilization_Ratio), and a target variable Attrition_Flag (indicating whether a customer churned or not).
	1.	Duplicate Check
I confirmed zero duplicates in the dataset (df.duplicated().sum() was 0). It was reassuring to see each row is unique.
	2.	Missing Values
Looking at df.isnull().sum(), I discovered that Education_Level and Marital_Status had missing entries (1,519 and 749 missing, respectively). This told me I either had to impute these values or treat “Unknown” as a valid category. I realized ignoring these might lose crucial data, so I decided to handle them carefully later.
	3.	Target Variable
Since Attrition_Flag was an object column with values like "Existing Customer" or "Attrited Customer", I converted it to numeric (0 for existing, 1 for churned) for modeling. I saw about 16.07% churn, so the dataset is indeed imbalanced. This step convinced me to keep an eye on advanced techniques (like SMOTE) to handle the minority churn class.
	4.	Client Identifiers
The dataset included CLIENTNUM, an identifier which I dropped for modeling (it doesn’t help predict churn and could risk data leakage).

Overall, my EDA allowed me to confirm I had the correct dataset, gauge its scale (about 1.6 MB in memory), and plan how to tackle missing or inconsistent values. I left this step feeling more confident about the data’s shape and any potential issues I might need to fix in the next phases.

2. Feature Distribution Analysis

Next, I zoomed in on the shapes and distributions of individual columns:

2.1 Numerical Features

I ran df[num_features].describe() on columns like Credit_Limit, Avg_Utilization_Ratio, Total_Trans_Amt, and Customer_Age. Some of my key observations:
	•	Credit_Limit: Ranged from roughly $1,438 up to $34,516, with a mean around $8,632 and a standard deviation near $9,089. A histogram revealed a skewed distribution, indicating a subset of customers had notably higher limits.
	•	Avg_Utilization_Ratio: Values near 0.0 up to 0.999, averaging 0.275. This ratio measures how much of their credit limit customers use—some might be near max usage, which can signal potential risk or high activity.
	•	Total_Trans_Amt: Ranged up to $18,484, with an average around $4,404. I recognized some “big spenders” could be outliers or legitimate heavy users.

I found boxplots particularly useful for spotting outliers. For instance, Total_Trans_Amt had a long tail, but in a bank scenario, that can be normal if a few high-net-worth individuals or heavy spenders drive that upper range.

2.2 Categorical Features

I examined columns like Income_Category, Education_Level, and Card_Category (if present). A countplot of Income_Category revealed categories such as “Less than $40K,” “$40K - $60K,” and “$120K+.” Interestingly, I also saw an “abc” or “Unknown” category, which means the data might have placeholders or small errors. If “abc” was 1,112 rows, I had to decide whether to treat it as missing or keep it as a separate group.

Proportions of each category told me which bracket was dominant. For example, “Less than $40K” was around 3,561 records, making it a big chunk of the dataset. Knowing the biggest categories can inform modeling decisions if I decide to do one-hot encoding or some special grouping.

2.3 Outliers

Through histograms and boxplots, I identified potential outliers in usage metrics. Instead of dismissing them, I reminded myself that in financial data, outliers can be legitimate. High-limit and high-transaction users might be valuable customers the bank can’t afford to lose.

Feature distribution analysis ultimately helped me plan for binning or capping certain columns and gave me a sense of how typical or atypical each variable’s range might be.

3. Correlation Analysis

After checking distributions, I computed df.corr() on numeric columns. I discovered:
	1.	Highly Correlated Pairs
For instance, Avg_Utilization_Ratio and Credit_Usage_Ratio had a correlation near 1.0. Also, Avg_Open_To_Buy correlated strongly with Credit_Limit (above 0.99). This indicated these pairs are nearly redundant, possibly measuring the same concept differently (like available credit vs. limit minus revolve).
	2.	Multicollinearity Considerations
In logistic regression or linear models, such strong correlations can inflate variance of coefficients. For tree-based models, it’s less of a problem, but I still noted that these features might not all be necessary.
	3.	Target Correlation
Although I sometimes convert Attrition_Flag to numeric 0/1 for correlation, a low correlation doesn’t imply a feature lacks predictive power, as churn might be non-linear. Still, if a feature has some moderate correlation with churn, it’s a useful signal.

Through a heatmap of correlation, I verified which sets of variables might be combined, dropped, or at least carefully interpreted. Correlation analysis is a quick but crucial check on data redundancy and potential collinearity traps.

4. Model Selection & Training

With the data well-understood and partially cleaned, I embarked on building predictive models for churn. My main steps:

4.1 Data Splitting

I did a train-validation-test split. For example, I might have around 56% in training (5,670 rows), 14% in validation (1,418 rows), and 30% in testing (3,039 rows). That final test set is essential for honest performance estimation.

4.2 Imbalanced Classes

Since only ~16% of customers churn, if I do nothing, a naive classifier can say “No churn” and be ~84% accurate. To handle this, I used:
	•	SMOTE: An oversampling strategy generating synthetic points in the feature space for churners.
	•	Undersampling: I considered it but found SMOTE more valuable, as it keeps all non-churn examples.

4.3 Choosing Models

I tried three main model types:
	1.	Logistic Regression: Great baseline, interpretable, and quick to train.
	2.	RandomForestClassifier: Captures non-linearities and interactions. Often a top performer in tabular data.
	3.	Neural Network: With Keras, potentially strong performance, but I noticed that if not carefully tuned, I might face issues like the network ignoring minority classes.

4.4 Hyperparameter Tuning

I employed:
	1.	RandomizedSearchCV on random forest with parameters like n_estimators (50, 100, 150), max_features (sqrt, log2), and min_samples_leaf (1, 2, 3). This found a best set of parameters with n_estimators=100, max_features='sqrt', etc.
	2.	Bayesian Optimization (using BayesSearchCV) to adaptively search a wider space (like max_depth, n_estimators up to 300). This yielded an even better random forest with an AUC of ~0.983 on test, which is remarkable.

Thus, I ended up with a few candidate final models: logistic regression as a baseline, a random forest, and a neural net (though the net’s recall for churners sometimes needed improvement).

5. Model Evaluation

5.1 Core Metrics
	•	Accuracy: My logistic regression achieved ~0.83 test accuracy, while the random forest soared to ~0.95, and a Bayesian-optimized forest even did ~0.95 or above.
	•	ROC-AUC: A key measure for ranking. Logistic regression was near 0.89–0.90, random forest reached ~0.90+ or even 0.98 if well-tuned.
	•	Precision, Recall, F1: Vital for churn. If I want to capture as many churners as possible, I check recall. If resources are limited for outreach, I focus on precision. The confusion matrix also matters: I want to keep false negatives low.

5.2 Confusion Matrix

For logistic regression, I might see something like:
	•	True Negatives: 2,120
	•	False Positives: 431
	•	False Negatives: 98
	•	True Positives: 390

Hence a decent recall for churners (390 / (390+98) ≈ 0.80) and a precision around 0.48 for churn. For random forests, recall and precision can be even higher, especially if the model is well-tuned and the data is oversampled for churners.

5.3 Overfitting Checks

I watch if a model’s training accuracy is extremely high but the validation or test accuracy is lower. Some random forests or neural nets can overfit if max_depth or the layer sizes are too large. Cross-validating on the validation set ensures I pick hyperparameters that generalize.

Overall, I was quite pleased to see the random forest approach yield high accuracy, recall, and an AUC in the 0.90+ range. This suggests we can effectively catch a big chunk of would-be churners.

6. Model Interpretation

6.1 Tree-Based Feature Importance

Random forests let me check model.feature_importances_. A bar chart might show:
	1.	Credit_Usage_Ratio or Avg_Utilization_Ratio
	2.	Total_Trans_Ct (transaction count)
	3.	Months_Inactive_12_mon
	4.	Contacts_Count_12_mon
	5.	Income_Category

as top features. This quickly tells me that usage patterns and activity levels strongly predict churn. People who seldom use their card or revolve too high a balance might be more likely to leave.

6.2 SHAP

I then used SHAP for deeper insight:
	•	A summary plot indicated how each feature shifts churn probability across the entire dataset. Maybe I found that Months_Inactive_12_mon has a large, positive SHAP value for many churners, or that higher Total_Trans_Ct lowers churn risk.
	•	Local SHAP for a single row let me see exactly which features push that row’s churn probability above or below the baseline. If Income_Category <= \$40K and Avg_Utilization_Ratio is high, that might lead to a big positive contribution to churn.

6.3 LIME

Using LIME, I can interpret a single prediction by approximating the model locally with a small interpretable model. For example, LIME might return:

('Total_Trans_Ct > 73.49', -0.25)
('Avg_Transactions_Active_Month > 2.28', -0.07)
('Contacts_Count_12_mon <= 2.00', -0.05)
('Months_Inactive_12_mon <= 2.00', -0.05)
('Total_Trans_Amt > 4576.00', +0.05)

A negative weight means it reduces churn probability, a positive weight means it increases it. So if Total_Trans_Ct > 73.49 is negative, that’s a sign the model thinks having more transactions means less churn risk. This local interpretation is beneficial for a retention officer dealing with one specific account.

6.4 Practical Significance

Interpretation fosters trust. Because banks want to see why a model labeled someone as high-churn risk, these local explanations are crucial. They also direct the bank to solutions: If the problem is inactivity, target re-engagement campaigns; if it’s high revolve usage, maybe offer a credit limit increase.

7. Conclusion & Recommendations

I’ll end with three major points:

7.1 Identify High-Risk Customers Early and Take Corrective Action

By building a robust churn pipeline—from EDA to feature engineering, SMOTE balancing, and random forest or logistic regression modeling—I can flag high-risk customers well before they leave. For instance, if the model’s churn probability for a certain account is 0.80, I can proactively trigger an outreach campaign, possibly saving that relationship. This is more cost-effective than reacquiring a lost customer.

7.2 Use Insights from Feature Importance to Improve Banking Services

If Credit_Usage_Ratio or Income_Category strongly influences churn, I might consider:
	1.	Fee Waivers: If low-income customers are prone to churn, removing or reducing their annual fee could reduce attrition.
	2.	Loyalty Rewards: If I see inactivity is a major factor, awarding extra points or offering deals after months of inactivity encourages usage.
	3.	Credit Limit Adjustments: If high revolve usage leads to churn, giving them a moderate credit line increase might boost satisfaction and reduce attrition.

By systematically checking which features drive churn, I can align my retention strategies with actual usage and demographic data.

7.3 Automate Customer Engagement Strategies Using AI-Driven Insights

Finally, it’s essential to deploy the model into the bank’s CRM. Each day or month, the system can:
	•	Score every customer’s churn likelihood.
	•	If churn_score > 0.7 (for example), automatically notify the retention team or auto-apply a loyalty perk.
	•	If Income_Category <= 1 (meaning less than $40K or $60K) and churn risk is high, a partial fee waiver might be instantly offered.
	•	If Avg_Utilization_Ratio > 0.5, the system could propose a credit limit bump to keep them satisfied.

This real-time integration uses advanced data science to shape day-to-day decisions, ensuring that Thera Bank remains proactive and keeps valuable customers, further improving revenue and loyalty.

Final Words (Student Reflection)

Through this entire seven-step project journey, I learned how thorough EDA sets the stage, ensuring I understand missing values, outliers, and imbalance. Feature distribution and correlation checks give me clarity on data patterns and redundancy. Then, model selection and training—with logistic regression, random forest, or neural networks—coupled with the right model evaluation (ROC-AUC, recall, confusion matrices) helps me pick the best approach. Interpretation via SHAP and LIME is critical for explaining each result, while the final conclusion ties everything into actionable business strategies for churn prevention. The entire pipeline exemplifies how data science can transform raw data into real, measurable impact for a bank’s bottom line.
""")
```
